```python
import agate
```

```python
attendance = agate.Table.from_csv('/Users/jackcorrigan/projects/council-flow/council-attendance/attendance.csv')
```

```python
print (attendance)
```

```python
grouped_alderman = attendance.group_by('alderman')
```

```python
count_alderman = grouped_alderman.aggregate([('count', agate.Count())])
```

```python
ordered_alderman = count_alderman.order_by('count', reverse=True)
```

```python
ordered_alderman.print_table(max_rows= 100)
```

```python
absence = attendance.group_by('status')
```

```python
absence_count = absence.aggregate([('count', agate.Count())])
```

```python
absence_count.print_table(max_rows=2)
```

```python
ald_committee = grouped_alderman.group_by(
    lambda row: row['committee'])
```

```python
ac_misses = ald_committee.aggregate([('count',agate.Count('status', False))])
```

```python
ac_misses_order = ac_misses.order_by('count', reverse=True)
```

```python
ac_misses_order.print_table(max_rows=300)
```

```python
ac_makes = ald_committee.aggregate([('count',agate.Count('status', True))])
```

```python
ac_makes_order = ac_makes.order_by('count', reverse=True)
```

```python
ac_makes_order.print_table(max_rows=300)
```

```python
marty_makes = ac_makes_order.where(lambda row: row['alderman'] == 'Quinn, Marty')
```

```python
marty_miss = ac_misses_order.where(lambda row: row['alderman'] == 'Quinn, Marty')
```

```python
marty_makes.print_table(max_rows=30)
```

```python
marty_miss.print_table(max_rows=30)
```
