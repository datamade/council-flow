```python
import agate
```

```python
attendance = agate.Table.from_csv('/Users/jackcorrigan/projects/council-flow/council-attendance/final_files/attendance_clean.csv')
```

```python
grouped_alderman = attendance.group_by('alderman')
```

```python
alderman_total_meetings = grouped_alderman.aggregate([('count', agate.Count())])
```

```python
ordered_alderman = alderman_total_meetings.order_by('count', reverse=True)
```

```python
ordered_alderman.print_table(max_rows= 100)
```

```python
column_names = ['Alderman', 'Total Meetings']
column_types = [agate.Text(), agate.Number()]
rows = ordered_alderman.rows
table = agate.Table(rows, column_names, column_types)
table.print_table(max_rows=100)
```

```python
alderman_attendance = grouped_alderman.aggregate([('Meetings_Attended', agate.Count('status',True)),('Total_Meetings', agate.Count())])
```

```python
alderman_attendance1 = alderman_attendance.where(lambda row: row['Total_Meetings'] > 50)
```

```python
def f(row):
    return row[1]/row[2]

aa_all = alderman_attendance1.compute([('Attendance %', agate.Formula(agate.Number(),f))
])

ordered_aa = aa_all.order_by('Attendance %', reverse=True)


table = ordered_aa.print_table(max_rows=100)
```
